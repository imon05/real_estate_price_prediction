```python
# Convert columns to numeric types
numeric_columns = ['total_sqft', 'bath', 'price', 'BHK']
df_2[numeric_columns] = df_2[numeric_columns].apply(pd.to_numeric, errors='coerce')

print(df_2.dtypes)  # To check the data types after conversion
print(df_2.head(10))
```

```python
# Engineer the price_per_sqft feature
df_2['price_per_sqft'] = (df_2['price'] * 100000) / df_2['total_sqft']

# Display the first few rows to check the new feature
df_2.head()
```

```python
# Grouping locations with frequency <= 10 into "Other"
location_counts = df_2['location'].value_counts()
df_2['location'] = df_2['location'].apply(lambda x: 'Other' if location_counts[x] <= 10 else x)
print(location_counts)
print(df_2['location'].value_counts())
```

```python
# Function to identify outliers using IQR
def detect_outliers_iqr(df_2, column):
    Q1 = df_2[column].quantile(0.25)
    Q3 = df_2[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (df_2[column] < lower_bound) | (df_2[column] > upper_bound)

# Apply outlier detection to relevant columns and drop rows with outliers
outlier_columns = ['total_sqft', 'bath', 'price']  # List of columns to check for outliers

# Create a boolean mask for outliers
outliers_mask = pd.Series([False] * len(df_2), index=df_2.index)  # Initialize all rows as not outliers

for col in outlier_columns:
    outliers = detect_outliers_iqr(df_2, col)
    print(f"Number of outliers in {col}: {outliers.sum()}")
    outliers_mask = outliers_mask | outliers  # Combine the masks

# Drop the outliers
df_3 = df_2[~outliers_mask]

# Check the shape of the dataset after removing outliers
print(f"Data shape after dropping outliers: {df_3.shape}")
```

```python
# Remove data points where total_sqft per BHK is less than 250
df_4 = df_3[~(df_3['total_sqft'] / df_3['BHK'] < 250)]

# Check the shape of the data after removal
print(f"Data shape after removing small room outliers: {df_4.shape}")
```

```python
def remove_price_per_sqft_outliers(df, threshold=1.5):  # Default is 1.5 std dev
    df_out = pd.DataFrame()
    for location, subdf in df.groupby('location'):
        mean = subdf['price_per_sqft'].mean()
        std_dev = subdf['price_per_sqft'].std()
        # Adjust threshold to 1.5 or 2 std deviations
        reduced_df = subdf[(subdf['price_per_sqft'] >= (mean - threshold * std_dev)) & (subdf['price_per_sqft'] <= (mean + threshold * std_dev))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

# Test with 1.5 standard deviations
df_5_1_5 = remove_price_per_sqft_outliers(df_4, threshold=1)
print(f"Data shape with 1 std deviation threshold: {df_5_1_5.shape}")
```

```python
# Group the data by location and BHK (number of rooms)
grouped = df_5_1_5.groupby(['location', 'BHK'])

# Calculate mean, median, and standard deviation of price_per_sqft for each group
price_stats = grouped['price_per_sqft'].agg(['mean', 'median', 'std', 'count']).reset_index()

# Display groups where count of properties is greater than 1 to check for variations
price_stats_with_variations = price_stats[price_stats['count'] > 1]

# Sort by location and BHK to easily spot differences
price_stats_with_variations = price_stats_with_variations.sort_values(by=['count'], ascending=False)

# Display the result
print(price_stats_with_variations)
```

```python
def remove_extreme_price_anomalies(df, threshold=0.3):
    df_out = pd.DataFrame()
    
    # Group by location
    for location, subdf in df.groupby('location'):
        subdf = subdf.sort_values('BHK')
        
        # Initialize a list to hold valid rows
        valid_rows = []
        prev_bhk = None
        prev_price_per_sqft = None
        
        # Loop through the sub-dataframe to compare price_per_sqft
        for index, row in subdf.iterrows():
            if prev_bhk is not None and row['BHK'] > prev_bhk:
                # Check if the current price_per_sqft is significantly lower than the previous one
                if row['price_per_sqft'] < prev_price_per_sqft * (1 - threshold):
                    continue  # Skip this row (anomaly detected)
            valid_rows.append(row)
            prev_bhk = row['BHK']
            prev_price_per_sqft = row['price_per_sqft']
        
        # Add the valid rows to the output dataframe
        df_out = pd.concat([df_out, pd.DataFrame(valid_rows)], ignore_index=True)
    
    return df_out

# Apply the function to remove extreme price anomalies
df_6 = remove_extreme_price_anomalies(df_5_1_5)

# Check the shape after removing anomalies
print(f"Data shape after removing extreme price anomalies: {df_6.shape}")

```

### Predicting Price of properties based on location, total area, BHK and number of bathrooms

```python
# Create a function to predict price 
def predict_price(location, sqft, bath, bhk):
    formatted_location = f"location_{location}"
    
    # Prepare the input feature vector (before scaling)
    feature_vector = np.zeros(len(x.columns))
    feature_vector[0] = sqft
    feature_vector[1] = bath
    feature_vector[2] = bhk

    # Check if the location exists in the columns
    if formatted_location in x.columns:
        loc_index = np.where(x.columns == formatted_location)[0][0]
        feature_vector[loc_index] = 1
    else:
        raise ValueError(f"Location '{formatted_location}' not found in the training data columns.")

    # Convert feature_vector to DataFrame with same column names as 'x'
    feature_vector_df = pd.DataFrame([feature_vector], columns=x.columns)

    # Scale the input features using the scaler
    feature_vector_scaled = scaler.transform(feature_vector_df)

    # Predict the log-transformed price using the scaled features
    log_prediction = model.predict(feature_vector_scaled)[0]

    # Reverse the log transformation to get the actual price
    actual_price = np.exp(log_prediction)

    return actual_price
```
